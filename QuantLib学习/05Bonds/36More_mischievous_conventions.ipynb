```python
import QuantLib as ql
import pandas as pd
```

```python
today = ql.Date(27, ql.January, 2011)
ql.Settings.instance().evaluationDate = today
```

```python
issueDate = ql.Date(28, ql.January, 2011)
maturity = ql.Date(31, ql.August, 2020)
schedule = ql.Schedule(issueDate, maturity, ql.Period(ql.Semiannual), ql.UnitedStates(ql.UnitedStates.GovernmentBond), ql.Unadjusted, ql.Unadjusted, ql.DateGeneration.Backward, False)

bond = ql.FixedRateBond(1, 100.0, schedule, [0.03625], ql.ActualActual(ql.ActualActual.Bond), ql.Unadjusted, 100.0)
```

```python
bond_yield = 0.034921
P1 = bond.dirtyPrice(bond_yield, bond.dayCounter(), ql.Compounded, ql.Semiannual)
```

```python
flat_curve = ql.FlatForward(bond.settlementDate(), bond_yield, ql.ActualActual(ql.ActualActual.Bond), ql.Compounded, ql.Semiannual)
engine = ql.DiscountingBondEngine(ql.YieldTermStructureHandle(flat_curve))
bond.setPricingEngine(engine)
P2 = bond.dirtyPrice()
```

```python
pd.DataFrame([(P1, P2)], columns=['with yield', 'with curve'], index=[''])
```

```python
pd.DataFrame([(ql.as_coupon(c).accrualStartDate(), ql.as_coupon(c).accrualEndDate()) for c in bond.cashflows()[:-1]], columns=('start date', 'end date'), index=range(1, len(bond.cashflows())))
```

```python
dayCounter = ql.ActualActual(ql.ActualActual.Bond)
T = dayCounter.yearFraction(ql.Date(28, ql.January, 2011), ql.Date(28, ql.February, 2011), ql.Date(28, ql.August, 2010), ql.Date(28, ql.February, 2011))
print(T)
```

```python
print(dayCounter.yearFraction(ql.Date(28, ql.January, 2011), ql.Date(28, ql.February, 2011), ql.Date(28, ql.February, 2010), ql.Date(28, ql.February, 2011)))
```

```python
y = ql.InterestRate(bond_yield, dayCounter, ql.Compounded, ql.Semiannual)
print(y.discountFactor(T))
```

```python
data = []
for i, c in enumerate(bond.cashflows()[:-1]):
    c = ql.as_coupon(c)
    A = c.amount()
    T = c.accrualPeriod()
    D = y.discountFactor(T)
    D_Cumulative = D if i == 0 else D * data[-1][3]
    A_discounted = A * D_Cumulative
    data.append((A, T, D, D_Cumulative, A_discounted))
data.append((100, '', '', D_Cumulative, 100*D_Cumulative))
data = pd.DataFrame(data, columns=('amount', 'T', 'discount', 'dicsount (cum.)', 'amount (disc.)'), index=['']*len(data))
data
```

```python
print(sum(data['amount (disc.)']))
```
